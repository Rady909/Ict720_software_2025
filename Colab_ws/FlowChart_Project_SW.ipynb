```python
import graphviz

# Function to generate the flowchart using Graphviz (Standard Style)
def generate_flowchart():
    dot = graphviz.Digraph(format='png')
    dot.attr(rankdir='TB', size='10')
    dot.attr('node', shape='box')

    # Define nodes
    dot.node('A', 'User (Mobile Device)')
    dot.node('B', 'Camera Capture (WebRTC API)')
    dot.node('C', 'Frontend (Web App - HTML, JS)')
    dot.node('D', 'Backend (Flask API)')
    dot.node('E', 'Ngrok (HTTPS Tunnel)')
    dot.node('F', 'Google Gemini API (AI Model)')
    dot.node('G', 'Database (Optional)')
    dot.node('H', 'Display Product Details (Web App)')

    # Define edges (flow connections)
    dot.edge('A', 'B', label='User scans product')
    dot.edge('B', 'C', label='Camera captures image')
    dot.edge('C', 'D', label='Send image to backend')
    dot.edge('D', 'E', label='Secure tunnel (Ngrok)')
    dot.edge('D', 'F', label='AI model detects product')
    dot.edge('F', 'D', label='Return product details')
    dot.edge('D', 'G', label='Store product info (optional)')
    dot.edge('D', 'H', label='Send details to frontend')
    dot.edge('H', 'A', label='User sees product info')

    return dot

# Function to generate a box-style diagram
def generate_box_diagram():
    dot = graphviz.Digraph(format='png')
    dot.attr(rankdir='TB', size='10')
    dot.attr('node', shape='box', style='filled', fillcolor='lightblue')

    # Define nodes
    dot.node('A', 'User (Mobile Device)', fillcolor='lightcoral')
    dot.node('B', 'Camera Capture (WebRTC API)', fillcolor='lightgreen')
    dot.node('C', 'Frontend (Web App - HTML, JS)', fillcolor='lightblue')
    dot.node('D', 'Backend (Flask API)', fillcolor='orange')
    dot.node('E', 'Ngrok (HTTPS Tunnel)', fillcolor='yellow')
    dot.node('F', 'Google Gemini API (AI Model)', fillcolor='violet')
    dot.node('G', 'Database (Optional)', fillcolor='gray')
    dot.node('H', 'Display Product Details (Web App)', fillcolor='cyan')

    # Define edges
    dot.edge('A', 'B', label='User scans product')
    dot.edge('B', 'C', label='Camera captures image')
    dot.edge('C', 'D', label='Send image to backend')
    dot.edge('D', 'E', label='Secure tunnel (Ngrok)')
    dot.edge('D', 'F', label='AI model detects product')
    dot.edge('F', 'D', label='Return product details')
    dot.edge('D', 'G', label='Store product info (optional)')
    dot.edge('D', 'H', label='Send details to frontend')
    dot.edge('H', 'A', label='User sees product info')

    return dot

# Generate and save the flowcharts
flowchart = generate_flowchart()
box_diagram = generate_box_diagram()

flowchart.render('product_recognition_flowchart', format='png', cleanup=False)
box_diagram.render('product_recognition_box_diagram', format='png', cleanup=False)

# Display the generated images in Colab
from IPython.display import display
from PIL import Image

print("Standard Flowchart:")
display(Image.open("product_recognition_flowchart.png"))

print("\nBox Diagram with Colors:")
display(Image.open("product_recognition_box_diagram.png"))
```
